# 1/Load the dataset into a data frame using Python


```python
#importation des bibliotheques
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
# Charger l'ensemble de données dans une trame de données
df = pd.read_csv('Video games (1).csv')
df.head()
```

# 2/ Clean the data as needed.


```python
df.info()
```

```python
df.isnull().sum()
```

```python
# from categorical variable into a number Using the label encoder function in the Sklearn library
#Encoder=LabelEncoder()
#df['Name']=Encoder.fit_transform(df['Name'])
#df['Platform']=Encoder.fit_transform(df['Platform'])
df['Publisher']=Encoder.fit_transform(df['Publisher'])
```

```python
#visualizing data one more time
df.info()
```

```python
# cleaning data
df['Year'].fillna(df['Year'].mean(),inplace=True)
df['Publisher'].fillna(df['Publisher'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

# 3/Create an area chart to show the trend of global sales over the years.


```python
fig = px.area(df, x='Year', y='Global_Sales')
fig
```

# 4/Create a histogram to show the distribution of global sales.


```python
#create the histogram
fig = px.histogram(df, x='Global_Sales', nbins=10)
fig
```

# Create a scatter plot to show the relationship between North American sales and European sales.



```python
# Tracer le graphique
#trace = go.Scatter(x=df['Area (ft.)'], y=df['Price'], mode='markers', marker=dict(color=df['Building Type']))
#data = [trace]
#layout = go.Layout(title='North American sales  vs European sales.',
                   #xaxis=dict(title='Year of sale'),
                   #yaxis=dict(title='Price'))
#fig = go.Figure(data=data, layout=layout)
#pyo.plot(fig)
# Step 5: Create a scatter plot to show the relationship between North American sales and European sales
plt.scatter(df['North American Sales'], df['European Sales'])
plt.xlabel('North American Sales')
plt.ylabel('European Sales')
plt.title('Relationship between North American Sales and European Sales')
plt.show()


```

6/Create a scatter plot to show the relationship between Japanese sales and Other sales.


```python
trace = go.Scatter(x=df['Area (ft.)'], y=df['Price'], mode='markers', marker=dict(color=df['Building Type']))
data = [trace]
layout = go.Layout(title='relationship between Japanese sales and Other sales',
                   xaxis=dict(title='Year of sale'),
                   yaxis=dict(title='Price'))
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```
